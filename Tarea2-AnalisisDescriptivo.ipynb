# EJERCICIO 1

Aplicar conceptos de análisis de datos descriptivos al siguiente conjunto de datos de ventas mensuales de una tienda minorista. La tabla incluya información de ventas mensuales durante un período de tiempo.

```python
import numpy as np
```

```python
#Datos anuales de ventas
ventas = [10000, 12000, 14000, 11000, 13000, 15000, 18000, 16000, 14000, 12000, 11000, 16000]
```

## Media

```python
#Calculo del promedio de ventas
mediaventas = np.mean(ventas)
print("Media en Ventas:", mediaventas)
```

## Mediana

```python
#Ordenamiento de los datos
sorted_ventas = np.sort(ventas)
print(sorted_ventas)
```

```python
#Calculo de la mediana en ventas anuales
medianaventas=np.median(sorted_ventas)
print("Mediana en Ventas:",medianaventas)
```

## Moda

```python
def calculate_mode(data):
    unique_values, counts = np.unique(data, return_counts=True)
    max_count = counts.max()
    mode_values = unique_values[counts == max_count]
    
    return mode_values
```

```python
data = np.array([10000, 12000, 14000, 11000, 13000, 15000, 18000, 16000, 14000, 12000, 11000, 16000])
```

```python
modaventas = calculate_mode(data)
print("Moda en Ventas:", modaventas)
```

## Varianza

```python
#La varianza es el promedio de las desviaciones
variance_ventas = np.var(data)
print("Varianza en Ventas:", variance_ventas)
```

## Desviación estándar

```python
std_dev_ventas = np.std(ventas)
print("Desviación Estándar:", std_dev_ventas)
```

## Gráfico de Variación de Ventas Anuales

```python
import matplotlib.pyplot as plt
meses = ['ENE','FEB','MAR', 'ABR', 'MAY', 'JUN', 'JUL', 'AGO', 'SEP', 'OCT', 'NOV', 'DIC']
```

```python
plt.bar(meses, ventas)
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.title('Ejemplo de Gráfico de Barras')
plt.show()
```

Se observa que el mes de mayor ventas en el anio es julio y el de menor ventas en enero. Me muestra una vista clara del resumen anual.

## Gráfico de distribucion mensual de ventas

```python
plt.hist(ventas, bins=8, edgecolor='k')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma de Datos de Ventas')
plt.show()
```

d) La tendencia es uniforme en el valor de las ventas lo que me permitiria establecer parametros para definir objetivos de ventas.
e) En la siguiente distribucion se observa que las ventas entre 11000 12000 14000 y 16000 USD son mas repetitivas lo que coincide con los valores arrojados en la Moda.
