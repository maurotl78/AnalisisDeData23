## Descripción del DataSet
Este Data set fue obtenido del sitio kaggle.com y me muestra un conjunto de datos que contiene 3.682 registros de cursos de 4 materias (Finanzas empresariales, Diseño gráfico, Instrumentos musicales y Diseño web) tomados de la plataforma de aprendizaje Udemy.
Fuente : https://www.kaggle.com/datasets/andrewmvd/udemy-courses

Se puede mostrar una información clara y no tan extensa donde se pueda aprovechar los conocimientos iniciales de Analisis de Datos.

## Cotenido
El contenido es: Id del Curso, Titulo del Curso, Url, Si es de pago, Precio, Numero de Suscriptiores, Numero de Opiniones, Numero de lectores, Nivel, Duracion, Fecha de Publicacion, Asunto o Grupo.
